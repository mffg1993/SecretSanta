```python
import smtplib
import random
import pandas as pd
import time as tm
from email.mime.text import MIMEText


# Automatization of the email sending
def send_email(subject, body, sender, recipients, password):
    msg = MIMEText(body)
    msg['Subject'] = subject
    msg['From'] = sender
    msg['To'] = ', '.join(recipients)
    with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp_server:
       smtp_server.login(sender, password)
       smtp_server.sendmail(sender, recipients, msg.as_string())
    print("Message sent to "+sender+"!")


# Definition of the body of the email
def text(giver,taker,Hints):
  msg="Hi "+giver+", \n \n I hope this email finds you well and ready for some holiday excitement! The moment has arrived to unveil the mystery of who you will be a Secret Santa for, and I'm here to spill the cocoa beans. \n \n ü•Å Drumroll, please... ü•Å\n \n You are in charge of getting a present to none other than "+taker+"! üéâ \n \n Although the job of finding the perfect gift is very hard, here are some hints that they kindly provided:\n \n"+Hints+"\n \n Get ready for a delightful surprise when the time comes for our gift exchange! Remember, the gift exchange is scheduled for December 10th in the SQO Christmas party, so mark your calendar and keep the festive spirit alive. Whether it's a small token or something more elaborate, the goal that the present should be around ____. If you have any questions feel free to reply to this email, and I'll make sure your message gets to the right festive elf. \n \n Wishing you a wonderful anticipation leading up to the big reveal! \n \n Best regards,\n \n A secret Elf"
  return(msg)


################################################################################################
# Does not discrimitate if the people are in a relationship (indicated in the list as pairs )
################################################################################################

def SecretSantaListV1(Npeople):
  #Containers of the Givers and Receivers
  G=[]
  R=[]

  # Select one person to initialize
  if kk==True:
    Gindex=0

  for ii in range(Npeople-1):

    # The giver is added to the list
    G.append(Gindex)

    # Select a random person from the list
    Rindex=random.randint(0,Npeople-1)

    #Is this person (giving a present to itself) or (are they already in the recepient list) or (Is it the first person)?
    while (Rindex==Gindex or Rindex in R or Rindex==G[0]):
      Rindex=random.randint(0,Npeople-1)

    # Updates the Giver
    Gindex=Rindex

    # Add to the list
    R.append(Rindex)

  # Finishes the arrays
  G.append(Gindex)
  R.append(G[0])

  #List of pairs
  Pairs=[ (G[ii],R[ii]) for ii in range(Npeople)]

  return(Pairs)





################################################################################################
# Generates a dearrengement for all the entries on a Dataframe
################################################################################################
def SuperSecretSanta(df):
  #Number of elements in the list
  Npeople=len(df)

  #Containers of the Givers and Receivers
  G=[]
  R=[]

  # Select one person to initialize
  Gindex=0

  for ii in range(Npeople-1):
    print(ii)

    # The giver is added to the list
    G.append(Gindex)

    # Select a random person from the list
    Rindex=random.randint(0,Npeople-1)

    #Is this person (giving a present to itself) or (are they already in the recepient list) or (Is it the first person) or is this person giving a present to their partner?
    while (Rindex==Gindex or Rindex in R or Rindex==G[0] or df['PartnerIndex'][Rindex]==df['PartnerIndex'][Gindex] ):
      Rindex=random.randint(0,Npeople-1)

    # Updates the Giver
    Gindex=Rindex

    # Add to the list
    R.append(Rindex)

  # Finishes the arrays
  G.append(Gindex)
  R.append(G[0])


  Pairs=[ (df['Name'][G[ii]],df['Email'][G[ii]],df['Name'][R[ii]],df['hints '][R[ii]]) for ii in range(Npeople)]
  return(Pairs)





```
